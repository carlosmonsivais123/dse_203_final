```python
import pandas as pd
from googlemaps import Client as GoogleMaps
import time
```

```python
df = pd.read_csv('housing_data_dirty.csv')
```

```python
df['address'] = df['address'].str.replace('Sales Office: ', '')
```

```python
gmaps = GoogleMaps('ENTER_API_KEY_HERE')
```

```python
df['long'] = ""
df['lat'] = ""
```

```python
for x in range(len(df['address'].values)):
    try:
        time.sleep(1) #to add delay in case of large DFs
        geocode_result = gmaps.geocode(df['address'][x])
        df['lat'][x] = geocode_result[0]['geometry']['location'] ['lat']
        df['long'][x] = geocode_result[0]['geometry']['location']['lng']
    except IndexError:
        print("Address was wrong...")
    except Exception as e:
        print("Unexpected error occurred.", e )
df.head()
```

```python
df2 = df[['price','beds','bath','sqft','lot_size','address','lat','long']]
```
