```python
import pandas as pd
```

```python
cen1 = pd.read_csv("gs://dse203_group_project/Data/Census/ACSDP5Y2020.DP05_2022-11-13T194924/ACSDP5Y2020.DP05-Data.csv",header=1)
cen2 = pd.read_csv('gs://dse203_group_project/Data/Census/ACSST5Y2020.S1903_2022-11-13T195021/ACSST5Y2020.S1903-Data.csv',header=1)
cen3 = pd.read_csv('gs://dse203_group_project/Data/Census/DECENNIALPL2020.P1_2022-11-13T194845/DECENNIALPL2020.P1-Data.csv',header=1)

cen1_cut = cen1[['Geography','Estimate!!Total housing units','Percent!!RACE!!Total population!!One race!!Black or African American','Percent!!RACE!!Total population!!One race!!Asian','Percent!!HISPANIC OR LATINO AND RACE!!Total population!!Hispanic or Latino (of any race)','Percent!!RACE!!Total population!!One race!!White']].copy()
cen2_cut = cen2[['Geography','Estimate!!Number!!HOUSEHOLD INCOME BY RACE AND HISPANIC OR LATINO ORIGIN OF HOUSEHOLDER!!Households','Estimate!!Number!!FAMILIES!!Families','Estimate!!Median income (dollars)!!HOUSEHOLD INCOME BY RACE AND HISPANIC OR LATINO ORIGIN OF HOUSEHOLDER!!Households']].copy()
cen3_cut = cen3[['Geography',' !!Total:']].copy()

cen1_cut.head()
```

```python
cen3.columns
```

```python
cen1_cut['Geo_id_clean'] = cen1_cut['Geography'].str[-5:].astype(int)
cen2_cut['Geo_id_clean'] = cen2_cut['Geography'].str[-5:].astype(int)
cen3_cut['Geo_id_clean'] = cen3_cut['Geography'].str[-5:].astype(int)

cen1_cut['GEOID'] = cen1_cut['Geography'].str[-11:]
cen2_cut['GEOID'] = cen2_cut['Geography'].str[-11:]
cen3_cut['GEOID'] = cen3_cut['Geography'].str[-11:]
```

```python
df_temp = pd.merge(cen1_cut, cen2_cut, how='inner')
df = pd.merge(df_temp, cen3_cut, how='inner')
```

```python
df.columns
```

```python
df.drop('Geography', axis=1,inplace=True)
df.rename(columns={'Geo_id_clean':'Geo_id','Estimate!!Total housing units':'Housing_unit',
       'Percent!!RACE!!Total population!!One race!!Black or African American':'Per_Black',
       'Percent!!RACE!!Total population!!One race!!Asian':'Per_Asian',
       'Percent!!HISPANIC OR LATINO AND RACE!!Total population!!Hispanic or Latino (of any race)':'Per_Hispanic',
       'Percent!!RACE!!Total population!!One race!!White':'Per_White',
       'Estimate!!Number!!HOUSEHOLD INCOME BY RACE AND HISPANIC OR LATINO ORIGIN OF HOUSEHOLDER!!Households':'Household_Income',
       'Estimate!!Number!!FAMILIES!!Families':'Number_Families',
       'Estimate!!Median income (dollars)!!HOUSEHOLD INCOME BY RACE AND HISPANIC OR LATINO ORIGIN OF HOUSEHOLDER!!Households':'Median_income',
       ' !!Total:':'Total_population'}, inplace=True)
df.head(20)
```

```python
df.tail(20)
```

```python
df = df[ ['Geo_id'] + [ col for col in df.columns if col != 'Mid' ] ]
```

```python
df.drop('Geo_Id', axis=1,inplace=True)
```

```python
df.to_csv('gs://dse203_group_project/Data/Census/clean_cen.csv')
```
