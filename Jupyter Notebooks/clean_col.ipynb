```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('with_coor.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df["beds"] = df["beds"].str.replace("bed", "")
df["beds"] = df["beds"].str.replace("Studio", "1")
df["beds"] = df["beds"].str.replace("\+", "")
```

```python
df["bath"] = df["bath"].str.replace("bath", "")
df["bath"] = df["bath"].str.replace("\+", "")
```

```python
df["price"] = df["price"].str.replace("\$", "")
df["price"] = df["price"].str.replace("\,", "")
df["price"] = df["price"].replace("Contact For Price", 0)
```

```python
df["sqft"] = df["sqft"].str.replace("\,", "")
df["lot_size"] = df["lot_size"].str.replace("\,", "")
```

```python
df['beds'] = df['beds'].astype(float)
df['bath'] = df['bath'].astype(float)
df['price'] = df['price'].astype(float)
df['sqft'] = df['sqft'].astype(float)
df['lot_size'] = df['lot_size'].astype(float)
```

```python
df.to_csv('clean_coor.csv')
```
